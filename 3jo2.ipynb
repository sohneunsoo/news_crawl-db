```python
# str -> 리스트로 형변환 ast
import ast
convertlist=[]
for i in range(2076):
    converted = ast.literal_eval(df['journalists'][i])
    convertlist.append(converted)
```

```python
type(convertlist)
```

```python
# 리스트로 형변환한걸 출력
text = convertlist
text
```

```python
# 기자정보를 딕셔너리로 만듦
dictlist=[]
for i in range(2076):
    companyname= dfnews.iloc[i]['company']
    for j in range(len(text[i])):
        if '=' in text[i][j]:
            a= text[i][j].split('=')[1].replace('기자','').replace('동아닷컴','').replace('한경닷컴','').replace('노컷뉴스','').replace('매경닷컴','')
        else:
            a= text[i][j].replace('기자','').replace('동아닷컴','').replace('한경닷컴','').replace('노컷뉴스','').replace('매경닷컴','')
        email = re.findall(r"[A-Za-z0-9_%+-.]+"
                 r"@[A-Za-z0-9.-]+"
                 r"\.[A-Za-z]{2,5}",a)
        korea_name = re.findall('[가-힣]+',a)
        if not email:
            continue

        for k in range(len(korea_name)):
            for y in range(len(email)):
                dictlist.append({'name':korea_name[k],
                                'email':email[y],
                                'company':companyname})
```

```python
# 위에 코드 출력
dictlist
```

```python
#길이
len(dictlist)
```

```python
# dictlist의 데이터 프레임
df= pd.DataFrame(dictlist)
```

```python
# 데이터프레임
df
```

```python
# 위 코드에 중복을 제거하니 순서가 비어서 reset_index를 사용함
journalist_df = df.drop_duplicates(['email']).reset_index()
```
